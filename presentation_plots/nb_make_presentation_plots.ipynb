```python
'''
[10-12-16]
+ Idea is to use $D2PIDIR_EXP(SIM)_E16/data_pcorr_011116 to make plots for presentation
+ Note that this data was made with 'd2pi_tree_BG' i.e. MM cut was not applied and
  therefore, I have to be mindful of that.
+ There may be other sublte nuances related to the fact that 'hvy' machinery is used
  to produce this data, but for the level of presentation not related to absolute observables,
  it should be OK.
  
+ Idea is to use DataFrames to keep code terse

+ Note currently only ER data is used and SR is ignored
'''
from __future__ import division

import os

import pandas as pd
import numpy as np 
from root_numpy import root2array, root2rec, tree2rec
from rootpy.plotting import Hist2D, Hist

NENTRIES=1000000000
#NENTRIES=10 #! use for debug

NDTYP=2
ER,SR=range(NDTYP)
DTYP_NAME=["ER","SR"]

FIN=[0 for i in range(NDTYP)]
ARR=[0 for i in range(NDTYP)]
DF=[0 for i in range(NDTYP)]

INDIR=[0 for i in range(NDTYP)]
INDIR[ER]=os.path.join(os.environ['D2PIDIR_EXP_E16'],"data_pcorr_011116")
INDIR[SR]=os.path.join(os.environ['D2PIDIR_SIM_E16'],"data_pcorr_011116")

FIN[ER]="%s/d2piR_wpcorr.root"%(INDIR[ER])
FIN[SR]="%s/d2piR_npcorr.root"%(INDIR[SR])
ARR[ER]=root2array(FIN[ER],'d2piR/tR',stop=NENTRIES)
#ARR[SR]=root2array(FIN[SR],'d2piR/tR',stop=NENTRIES)
DF[ER] = pd.DataFrame(ARR[ER])
#DF[SR] = pd.DataFrame(ARR[SR])

OUTDIR=os.environ['PRESENTATION_PLOTS_DATADIR']
#print "OUTDIR=",OUTDIR

#! Q2,W for thesis
Q2MIN,Q2MAX=2.0,5.0
Q2BINL=[2.0,2.4,3.0,3.5,4.2,5.0]
WMIN,WMAX,WBINW=1.400,2.125,0.025
WBINL=np.arange(WMIN,WMAX+WBINW,WBINW)

#! MM cut
MM2_MIN=-0.04
MM2_MAX=0.06

#! Setup  basic cuts on DF[ER]: cuts on MM and Q2,W
CUT_MM  =(DF[ER]['mm2ppip']>MM2_MIN) & (DF[ER]['mm2ppip']<MM2_MAX)
CUT_Q2_W=(DF[ER]['Q2']>=Q2MIN) & (DF[ER]['Q2']<Q2MAX) & (DF[ER]['W']>=WMIN) & (DF[ER]['W']<WMAX)
CUT_BASIC=CUT_MM & CUT_Q2_W
```
