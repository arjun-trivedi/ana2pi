```python
wsel_uel=[1.425,1.575,1.725,1.850,2.000,2.125,2.275,2.425,2.550,2.700,2.825,2.950,3.000]
binw_fin=[]
for x in w_uel:
    if x in wsel_uel:
        binw_fin.append( ((x-Mpi)-Mppi)/14 )
    else:
        binw_fin.append(0)
#plt.scatter(w_uel,binw)
last_idx=-1
for x in binw_fin:
    if x!=0:
        idx=binw_fin.index(x)
        n=idx-(last_idx+1)
        #print idx,n
        binw_fin[last_idx+1:idx]=[x for y in range(n)]
        last_idx=idx
# print binw
# print [round(x,2) for x in binw ]
```

```python
fig=plt.figure(figsize=(18,8))
ax=plt.subplot(111)
ax.scatter(w_uel,binw_fin)
ax.set_xlabel("WBINS_UE [GeV]")
ax.set_xticks(w_uel)
plt.xticks(rotation='vertical')
ax.set_ylabel("M binw [MeV]")
#! mark upper edges of selected w bins
ymin=ax.get_ylim()[0]+0.01
ymax=ax.get_ylim()[1]-0.01
for i in wsel_uel:
    ax.vlines(i,ymin,ymax)
plt.grid()
```
