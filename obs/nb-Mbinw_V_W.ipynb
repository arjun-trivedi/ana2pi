```python
from __future__ import division
import atlib
wbng=atlib.init_bng((1.3,3,0.025),3)
#w_uel=wbng['BINS_UE']
w=wbng['BINS']

w_crs=[1.300,1.425,1.575,1.725,1.850,2.000,2.125,2.275,2.425,2.550,2.700,2.825,2.950,3.000]

Mppi=round(0.938 + 0.139,3)
Mpi=0.139
binw=[((x-Mpi)-Mppi)/14 for x in w]
binw_r2=[round(x,2) for x in binw]
binw_r3=[round(x,3) for x in binw]

fig=plt.figure(figsize=(18,8))

ax_r2=plt.subplot(211)
ax_r2.scatter(w,binw_r2)
# ax_r2.set_xlabel("WBINS_UE [GeV]")
ax_r2.set_xticks(w)
plt.xticks(rotation='vertical')
#plt.setp( ax_r2.get_xticklabels(), visible=False)
ax_r2.set_ylabel("M binw (r2) [MeV]")
plt.grid()
#! mark upper edges of selected w bins
ymin=ax_r2.get_ylim()[0]+0.01
ymax=ax_r2.get_ylim()[1]-0.01
for i in w_crs:
    ax_r2.vlines(i,ymin,ymax)

ax_r3=plt.subplot(212,sharex=ax_r2)
ax_r3.scatter(w,binw)#_r3)
ax_r3.set_xlabel("W [GeV]")
ax_r3.set_xticks(w)
plt.xticks(rotation='vertical')
ax_r3.set_ylabel("M binw [MeV]")
#! mark upper edges of selected w bins
# ymin=ax_r3.get_ylim()[0]+0.01
# ymax=ax_r3.get_ylim()[1]-0.01
for i in w_crs:
    ax_r3.vlines(i,ymin,ymax)
plt.grid()
```

```python
binw_fin=[]
for x in w:
    if x in w_crs:
        binw_fin.append( ((x-Mpi)-Mppi)/14 )
    else:
        binw_fin.append(0)
#plt.scatter(w_uel,binw)
last_idx=-1
for x in binw_fin:
    if x!=0:
        idx=binw_fin.index(x)
        n=idx-(last_idx+1)
        #print idx,n
        binw_fin[last_idx+1:idx]=[x for y in range(n)]
        last_idx=idx
# print binw
# print [round(x,2) for x in binw ]
```

```python
fig=plt.figure(figsize=(18,8))
ax=plt.subplot(111)
ax.scatter(w,binw_fin)
ax.set_xlabel("WBINS_UE [GeV]")
ax.set_xticks(w)
plt.xticks(rotation='vertical')
ax.set_ylabel("M binw [MeV]")
#! mark upper edges of selected w bins
ymin=ax.get_ylim()[0]+0.01
ymax=ax.get_ylim()[1]-0.01
for i in w_crs:
    ax.vlines(i,ymin,ymax)
plt.grid()
```
