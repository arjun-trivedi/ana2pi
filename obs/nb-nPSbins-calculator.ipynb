#### Plot numer of Phase Space(PS) bins vs. W 
+ Use VST1 variables

```python
Mpi=0.139
Mp=0.938
Mppip_th=1.1
Mpippim_th=0.2780
w=np.linspace(1.3,3.0,68)
nbins_Mppip=[((x-Mpi)-Mppip_th)/0.025 for x in w]
nbins_Mpippim=[((x-Mp)-Mpippim_th)/0.025 for x in w]
nbins=[M1*M2*10*10*10 for M1,M2 in zip(nbins_Mppip,nbins_Mpippim)]
fig=plt.figure(figsize=(8,10))
plt.plot(w,nbins)
plt.grid()
plt.semilogy()
plt.xlabel("W [GeV]")
plt.ylabel("nbins")
plt.title("Number of bins in PS vs. W (VST1)")
```
