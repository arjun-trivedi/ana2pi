#### Calculate number of Phase Space(PS) bins vs. W and plot it vs. W and Q2,W
+ Use VST1 variables

```python
"""
Calculate nPSbins vs. W and plot vs. W
"""
from __future__ import division
M=1000000
Mpi=0.139
Mp=0.938
Mppip_th=1.1
Mpippim_th=0.2780
#w=np.linspace(1.3,3.0,68)
#np.around(w,3)
wmin=1.0
wmax=3.0
wbinw=0.025
nwbins=int(round((wmax-wmin)/wbinw,0))
w=[wmin+(i*wbinw) for i in np.arange(nwbins)]
nbins_Mppip=[((x-Mpi)-Mppip_th)/0.022 for x in w]
nbins_Mpippim=[((x-Mp)-Mpippim_th)/0.022 for x in w]
nbins_Mppip1=[((x-Mpi)-Mppip_th)/0.035 for x in w]
nbins_Mpippim1=[((x-Mp)-Mpippim_th)/0.035 for x in w]
nbins_Mppip2=[14 for x in w]
nbins_Mpippim2=[14 for x in w]
nbins=[(nbins_M1*nbins_M2*10*10*10)/M for nbins_M1,nbins_M2 in zip(nbins_Mppip,nbins_Mpippim)]
nbins1=[(nbins_M1*nbins_M2*10*10*10)/M for nbins_M1,nbins_M2 in zip(nbins_Mppip1,nbins_Mpippim1)]
nbins2=[(14*14*10*10*10)/M for M1,M2 in zip(nbins_Mppip2,nbins_Mpippim2)]
fig=plt.figure(figsize=(20,10))
fig.suptitle("Number of bins in PS vs. W (VST1)",fontsize=14, fontweight='bold')

ax=plt.subplot(211)
ax.plot(w,nbins,label="M-bng=22 MeV/bin")
ax.plot(w,nbins1,label="M-bng=35 MeV/bin")
ax.plot(w,nbins2,label="M-bng=Golovach (14 bins)")
ax.grid()
ax.semilogy()
ax.set_xticks(w)#rotation='vertical')
plt.xticks(rotation='vertical')
ax.tick_params(labelsize='large')
ax.set_xlabel("W [GeV]")
ax.set_ylabel("nbins [M]")
ax.set_title("Logarithmic scale")
plt.legend(loc='upper left')

plt.subplot(212)
plt.plot(w,nbins,label="M bng=22 MeV/bin")
plt.plot(w,nbins1,label="M bng=35 MeV/bin")
plt.plot(w,nbins2,label="M-bng=Golovach (14 bins)")
plt.xticks(w,rotation='vertical')
plt.grid()
plt.tick_params(labelsize='large')
plt.xlabel("W [GeV]")
plt.ylabel("nbins [M]")
plt.title("Linear scale")
plt.legend(loc='upper left')
```

```python
nbins_M1M2=[nbins_M1*nbins_M2 for nbins_M1,nbins_M2 in zip(nbins_Mppip,nbins_Mpippim)]

ax1=plt.subplot(211)
plt.plot(w,nbins_Mppip)
plt.plot(w,nbins_Mpippim)
ax1.set_ylabel("nbins-M1/M2")

ax2=plt.subplot(212,sharex=ax1)
plt.plot(w,nbins_M1M2)
ax2.set_xlabel("W [GeV]")
ax2.set_ylabel("nbins-M1*nbins-M2")
```

```python
"""
Plot nPSbins(Q2,W)
"""
import rootnotes

#! Create nbins_vs_w
nbins_vs_w=zip(w,nbins)
#print nbins_vs_w

#! Create histogram
wbins,wmin,wmax=80,1.0,3.0 #25 MeV/bin
q2bins,q2min,q2max=7,1.2,5.0 # place-holder bng, for it will be modified by variable bng
q2_varbins=array("d",[1.0,1.2,1.6,2.0,2.4,3.0,3.5,4.2,5.0])
h=ROOT.TH2F("h","h",wbins,wmin,wmax,q2bins,q2min,q2max)
h.GetYaxis().Set(len(q2_varbins)-1,q2_varbins)
h.SetXTitle("W [GeV]")
h.SetYTitle("Q^2 [GeV^2]")
print("bins for h=%d,%d"%(h.GetXaxis().GetNbins(), h.GetYaxis().GetNbins()))

#! Fill h
for q2 in q2_varbins:
    for id,d in enumerate(nbins_vs_w):
        w=nbins_vs_w[id][0]
        nbins=nbins_vs_w[id][1]
        #print q2,w,nbins
        binx=h.GetXaxis().FindBin(w+(0.025/2));
        biny=h.GetYaxis().FindBin(q2);
        bin=h.GetBin(binx,biny)
        h.SetBinContent(bin,nbins)
c=rootnotes.default_canvas()
ROOT.gStyle.SetOptStat(0)
h.Draw("colz")
c
```
