### Determining Coarse-W bins for my analysis

* The h8s need to be made in Coarse W-bins(which is just a grouping of the Analysis W-bins)
* The Course-W bins need to be decided based on how they affect the bin width in M(Mppi,Mpipi) (nbins-M=14; as per Golovach)

```python
"""
In this block, I show how M-binw depends on the Analysis W-bins
"""

from __future__ import division
import atlib

#! Obtain the analysis bins
wbng=atlib.init_bng((1.3,3,0.025),3)
wbins=wbng['BINS']

Mppi=round(0.938 + 0.139,3)
Mpipi=round(0.139 + 0.139,3)
Mpi=0.139
Mp=0.938
NMbins=14

binw_Mppi=[((x-Mpi)-Mppi)/NMbins for x in wbins]
binw_Mpipi=[((x-Mp)-Mpipi)/NMbins for x in wbins]

fig=plt.figure(figsize=(18,4))
ax=plt.subplot(111)
ax.scatter(wbins,binw_Mppi,color='r',marker=">",label="Mppi")
ax.scatter(wbins,binw_Mpipi,color='g',marker="o",alpha=0.5,label="Mpipi")
ax.set_xlabel("W [GeV]")
ax.set_xticks(wbins)
plt.xticks(rotation='vertical')
ax.set_ylabel("binw [GeV]")
plt.grid()
plt.legend(loc="upper left")
```

```python
"""
In this block, I round the M-binw to 2 decimal places and the following Course-W bins naturally fall out
on visually inspecting the plot.
+ wcrsbins=[1.300,1.425,1.575,1.725,1.850,2.000,2.125,2.275,2.425,2.550,2.700,2.825,2.950,3.000]
"""
#! The following Course-W bins was selected after viewing the graph generated in this block
wcrsbins=[1.300,1.425,1.575,1.725,1.850,2.000,2.125,2.275,2.425,2.550,2.700,2.825,2.950,3.000]

binw_Mppi_r2=[round(x,2) for x in binw_Mppi]
binw_Mpipi_r2=[round(x,2) for x in binw_Mpipi]

fig=plt.figure(figsize=(18,8))
ax_r2=plt.subplot(211)
ax_r2.scatter(wbins,binw_Mppi_r2,color='r',marker=">",label="Mppi")#,alpha=0.5)
ax_r2.scatter(wbins,binw_Mpipi_r2,color='g',marker="o",alpha=0.5,label="Mpipi")
# ax_r2.set_xlabel("WBINS_UE [GeV]")
ax_r2.set_xticks(wbins)
plt.xticks(rotation='vertical')
#plt.setp( ax_r2.get_xticklabels(), visible=False)
ax_r2.set_ylabel("binw (r2) [GeV]")
plt.grid()
#! mark upper edges of selected w bins
ymin=ax_r2.get_ylim()[0]+0.01
ymax=ax_r2.get_ylim()[1]-0.01
for i in wcrsbins:
    ax_r2.vlines(i,ymin,ymax)
plt.legend(loc="upper left")

#! For ease of comparing rounded M-binw, plot the raw M-binw vs. w-bin again
ax=plt.subplot(212,sharex=ax_r2)
ax.scatter(wbins,binw_Mppi,color='r',marker=">",label="Mppi")
ax.scatter(wbins,binw_Mpipi,color='g',marker="o",alpha=0.5,label="Mpipi")
ax.set_xlabel("W [GeV]")
ax.set_xticks(wbins)
plt.xticks(rotation='vertical')
ax.set_ylabel("binw [GeV]")
for i in wcrsbins:
    ax.vlines(i,ymin,ymax)
plt.grid()
plt.legend(loc="upper left")
```

```python
"""
The following code is a test of the algorith in my C++ code that will be used to calculate M-bng depending
on the Course-W bin
"""
binw_Mppi_fin=[]
binw_Mpipi_fin=[]
for x in wbins:
    if x in wcrsbins:
        binw_Mppi_fin.append( ((x-Mpi)-Mppi)/NMbins )
        binw_Mpipi_fin.append( ((x-Mp)-Mpipi)/NMbins )
    else:
        binw_Mppi_fin.append(0)
        binw_Mpipi_fin.append(0)

last_idx=-1
for x in binw_Mppi_fin:
    if x!=0:
        idx=binw_Mppi_fin.index(x)
        n=idx-(last_idx+1)
        #print idx,n
        binw_Mppi_fin[last_idx+1:idx]=[x for y in range(n)]
        last_idx=idx

last_idx=-1
for x in binw_Mpipi_fin:
    if x!=0:
        idx=binw_Mpipi_fin.index(x)
        n=idx-(last_idx+1)
        #print idx,n
        binw_Mpipi_fin[last_idx+1:idx]=[x for y in range(n)]
        last_idx=idx
```

```python
fig=plt.figure(figsize=(18,8))
ax=plt.subplot(111)
ax.scatter(wbins,binw_Mppi_fin,color='r',marker=">",label="Mppi")
ax.scatter(wbins,binw_Mpipi_fin,color='g',marker="o",alpha=0.5,label="Mpipi")
ax.set_xlabel("WBINS_UE [GeV]")
ax.set_xticks(wbins)
plt.xticks(rotation='vertical')
ax.set_ylabel("binw [MeV]")
#! mark upper edges of selected w bins
ymin=ax.get_ylim()[0]+0.01
ymax=ax.get_ylim()[1]-0.01
for i in wcrsbins:
    ax.vlines(i,ymin,ymax)
plt.grid()
plt.legend(loc="upper left")
```
