```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
from rootpy.plotting import Hist
from rootpy.plotting import root2matplotlib as rplt
from rootpy.interactive import wait
from scipy.stats import norm
import matplotlib.mlab as mlab
import ROOT

def gauss(v, par):
    arg = 0;
    #binw=((0.02-(-0.02))/100)
    if (par[2] != 0): arg = (v[0] - par[1])/par[2];
    fitval = par[0]*(1/(sqrt(2*pi)*par[2]))*exp(-0.5*arg*arg)
    return fitval

h=Hist(100,-4,4)
x1=np.random.normal(0,1,1000000)
h.fill_array(x1)
ax=plt.subplot(111)
n,bins,patches=rplt.hist(h,ax)
ax.set_ylim(ax.get_ylim()[0],ax.get_ylim()[1])
ax.vlines(0.0,ax.get_ylim()[0],ax.get_ylim()[1],colors="red")

ROOT.gStyle.SetOptFit(1111)
n=h.GetEntries()
print n
scale= 1/(n*.08)
print scale
c=ROOT.TCanvas()
h.Scale(scale)
#h.Scale((1/n),"width")
#h.Scale(0.5)
fgauss=ROOT.TF1("fgauss",gauss,-5,5,3)
fgauss.SetParameters(1,0,1)
fgauss.FixParameter(0,1)
h.Fit("fgauss")
#h.Fit("gaus")
h.Draw()

if not ROOT.gROOT.IsBatch():
		plt.show()
	 	# wait for you to close the ROOT canvas before exiting
	 	wait(True)



```

```python
"""
Test Plotting and Fitting a histogram using
1. matplotlib & scipy
2. rootpy.plotting
"""
from __future__ import division
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.mlab as mlab

from rootpy.plotting import Hist
from rootpy.plotting import root2matplotlib as rplt

NSAMPLES=10000
XMIN=-5
XMAX=5
NBINS=100

x=np.random.normal(0,1,NSAMPLES)

# Plot Histogram & Fit: Method 1.
plt.figure(figsize=(15,10))
ax1=plt.subplot(221)
n1,bins1,patches1=ax1.hist(x,NBINS,range=(XMIN,XMAX))#,normed=True)
ax1.set_ylim(min(n1),max(n1))
(mu1,sigma1) = norm.fit(x)
y1 = mlab.normpdf(bins1,mu1,sigma1)
plt.plot(bins1,y1,'r--',linewidth=2)
plt.title(r'$\mu=%.3f,\ \sigma=%.3f$' %(mu1, sigma1))

# Plot Histogram: Method 2.
h=Hist(NBINS,XMIN,XMAX)
h.fill_array(x)
ax2=plt.subplot(222)
n2,bins2,patches2=rplt.hist(h,ax2)
ax2.set_ylim(h.GetMinimum(),h.GetMaximum())

# Plot Histogram with bin content normed to pdf: Method 1.
ax3=plt.subplot(223)
n3,bins3,patches3=ax3.hist(x,NBINS,range=(XMIN,XMAX),normed=True)
ax3.set_ylim(min(n3),max(n3))

# Plot Histogram with bin content normed to pdf: Method 2.
ax4=plt.subplot(224)
binw=(5-(-5))/10
scale=(1/(h.GetEntries()*binw))
h.Scale(scale)
n4,bins4,patches4=rplt.hist(h,ax4)
ax4.set_ylim(h.GetMinimum(),h.GetMaximum())

plt.show()
```
