```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
from rootpy.plotting import Hist
from rootpy.plotting import root2matplotlib as rplt
from rootpy.interactive import wait
from scipy.stats import norm
import matplotlib.mlab as mlab
import ROOT

def gauss(v, par):
    arg = 0;
    #binw=((0.02-(-0.02))/100)
    if (par[2] != 0): arg = (v[0] - par[1])/par[2];
    fitval = par[0]*(1/(sqrt(2*pi)*par[2]))*exp(-0.5*arg*arg)
    return fitval

h=Hist(100,-4,4)
x1=np.random.normal(0,1,10000)
h.fill_array(x1)
ax=plt.subplot(111)
n,bins,patches=rplt.hist(h,ax)
ax.set_ylim(ax.get_ylim()[0],ax.get_ylim()[1])
ax.vlines(0.0,ax.get_ylim()[0],ax.get_ylim()[1],colors="red")

ROOT.gStyle.SetOptFit(1111)
n=h.GetEntries()
print n
scale= 1/(n*.08)
print scale
c=ROOT.TCanvas()
h.Scale(scale)
#h.Scale((1/n),"width")
#h.Scale(0.5)
fgauss=ROOT.TF1("fgauss",gauss,-5,5,3)
fgauss.SetParameters(1,0,1)
fgauss.FixParameter(0,1)
h.Fit("fgauss")
#h.Fit("gaus")
h.Draw()

if not ROOT.gROOT.IsBatch():
		plt.show()
	 	# wait for you to close the ROOT canvas before exiting
	 	wait(True)



```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.mlab as mlab

x=np.random.normal(0,1,1000000)
n,bins,patches=plt.hist(x,100,normed=True)
(mu,sigma) = norm.fit(x)
y = mlab.normpdf(bins,mu,sigma)
plt.plot(bins,y,'r--',linewidth=2)
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
```
