```python
from __future__ import division
import matplotlib.pyplot as plt
import numpy as np
from rootpy.plotting import Hist
from rootpy.plotting import root2matplotlib as rplt
from rootpy.interactive import wait
from scipy.stats import norm
import matplotlib.mlab as mlab
import ROOT

def gauss(v, par):
    arg = 0;
    #binw=((0.02-(-0.02))/100)
    if (par[2] != 0): arg = (v[0] - par[1])/par[2];
    fitval = par[0]*(1/(sqrt(2*pi)*par[2]))*exp(-0.5*arg*arg)
    return fitval

h=Hist(100,-4,4)
x1=np.random.normal(0,1,1000000)
h.fill_array(x1)
ax=plt.subplot(111)
n,bins,patches=rplt.hist(h,ax)
ax.set_ylim(ax.get_ylim()[0],ax.get_ylim()[1])
ax.vlines(0.0,ax.get_ylim()[0],ax.get_ylim()[1],colors="red")

ROOT.gStyle.SetOptFit(1111)
n=h.GetEntries()
print n
scale= 1/(n*.08)
print scale
c=ROOT.TCanvas()
h.Scale(scale)
#h.Scale((1/n),"width")
#h.Scale(0.5)
fgauss=ROOT.TF1("fgauss",gauss,-5,5,3)
fgauss.SetParameters(1,0,1)
fgauss.FixParameter(0,1)
h.Fit("fgauss")
#h.Fit("gaus")
h.Draw()

if not ROOT.gROOT.IsBatch():
		plt.show()
	 	# wait for you to close the ROOT canvas before exiting
	 	wait(True)



```

```python
"""
Test Plotting and Fitting a norm-hist to norm-pdf using
1. matplotlib & scipy
2. rootpy.plotting & ROOT::TF1

where norm implies that each bin is so normalized such that the 
histogram bin contents reflect the pdf at the bin-center:
    * bincontent=bincontent/(n*binw); where n = Total Entries
        + This is directly implemented with matplotlib.hist.normed=True
        + For rootpy histogram, I have to manually scale the histogram by (1/(n*binw))
"""
from __future__ import division
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import norm
import matplotlib.mlab as mlab

from rootpy.plotting import Hist
from rootpy.plotting import root2matplotlib as rplt
import ROOT

NSAMPLES=10000
XMIN=-5
XMAX=5
NBINS=100

def gauss_pdf(x, par):
    fitval = ROOT.TMath.Gaus(x[0],par[0],par[1],True)
    return fitval
fgauss_pdf=ROOT.TF1("fgauss_pdf",gauss_pdf,XMIN,XMAX,2)
ROOT.gStyle.SetOptFit(1111)

#c=ROOT.TCanvas("test_gauss_pdf","test_gauss_pdf")
#ftest_gauss_pdf=fgauss_pdf
#ftest_gauss_pdf.SetParameters(0,1)
#ftest_gauss_pdf.Draw()

CWIDTH=300
CHEIGHT=300


x=np.random.normal(0,1,NSAMPLES)
b=np.random.uniform(XMIN,XMAX,1000)
#t=x
t=np.concatenate((x,b))

# Plot matplotlib-hist & pdf-scipy-Fit
plt.figure(figsize=(15,10))
ax1=plt.subplot(221)
ax1.set_title("matplotlib-hist & pdf-scipy-Fit")
n1,bins1,patches1=ax1.hist(t,NBINS,range=(XMIN,XMAX))#,normed=True)
ax1.set_ylim(min(n1),max(n1))
(mu1,sigma1) = norm.fit(t)
y1 = mlab.normpdf(bins1,mu1,sigma1)
plt.plot(bins1,y1,'r--',linewidth=2)
plt.title('matplotlib-hist & pdf-scipy-Fit: $\mu=%.3f,\ \sigma=%.3f$' %(mu1, sigma1))

# Plot rootpy.plotting-hist & pdf-TF1-Fit
h=Hist(NBINS,XMIN,XMAX)
h.fill_array(t)
ax2=plt.subplot(222)
ax2.set_title("rootpy.plotting-hist & pdf-TF1-Fit")
n2,bins2,patches2=rplt.hist(h,ax2)
ax2.set_ylim(h.GetMinimum(),h.GetMaximum())
c2=ROOT.TCanvas("c2","rootpy.plotting-hist & pdf-TF1-Fit",CWIDTH,CHEIGHT)
fgauss_pdf.SetParameters(0,1)
h.Fit("fgauss_pdf","L")
#h.Fit("gaus")
h.Draw()

# Plot matplotlib-norm-hist & pdf-scipy-Fit.
ax3=plt.subplot(223)
ax3.set_title("matplotlib-norm-hist & pdf-scipy-Fit")
n3,bins3,patches3=ax3.hist(t,NBINS,range=(XMIN,XMAX),normed=True)
ax3.set_ylim(min(n3),max(n3))
(mu3,sigma3) = norm.fit(t)
y3 = mlab.normpdf(bins3,mu3,sigma3)
plt.plot(bins3,y3,'r--',linewidth=2)
plt.title('matplotlib-norm-hist & pdf-scipy-Fit:$\mu=%.3f,\ \sigma=%.3f$' %(mu3, sigma3))

# Plot rootpy.plotting-norm-hist & pdf-TF1-Fit
ax4=plt.subplot(224)
ax4.set_title("rootpy.plotting-norm-hist & pdf-TF1-Fit")
binw=(XMAX-XMIN)/NBINS
scale=(1/(h.GetEntries()*binw))
h.Scale(scale)
n4,bins4,patches4=rplt.hist(h,ax4)
ax4.set_ylim(h.GetMinimum(),h.GetMaximum())
c4=ROOT.TCanvas("c4","rootpy.plotting-norm-hist & pdf-TF1-Fit",CWIDTH,CHEIGHT)
fgauss_pdf.SetParameters(0,1)
h.Fit("fgauss_pdf","L")
h.Draw()
h.Fit("gaus","0+")
f=h.GetFunction("gaus")
f.SetLineColor(ROOT.gROOT.ProcessLine("kGreen"))
f.Draw("same")
c4.Update()


plt.show()
```
