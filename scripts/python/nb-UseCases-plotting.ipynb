```python
"""
Use Case for:
1. making graphs/plots with 'rootpy.plotting' 
2. plotting them `PyROOT` style i.e. using TCanvas
"""
from matplotlib import pyplot as plt
from rootpy.plotting import Hist2D, Hist
import numpy as np
import ROOT 
from rootpy.interactive import wait

print __doc__
h2 = Hist2D(100, -3, 3, 100, -3, 3)
h2.fill_array(np.random.randn(1000000, 2))
ch2=ROOT.TCanvas("h2","h2")
h2.draw("colz")

h=Hist(100,-10,10)
h.fill_array(np.random.normal(2,14,1000))
ch1=ROOT.TCanvas("h","h")
h.draw()

if not ROOT.gROOT.IsBatch():
    plt.show()
    # wait for you to close the ROOT canvas before exiting
    wait(True)
```

```python
"""
Use Case for:
1. Making graphs/plots with 'rootpy.plotting' 
2. Plotting them using `matplotlib`
"""
from matplotlib import pyplot as plt
from rootpy.plotting import Hist2D, Hist
import numpy as np

print __doc__
a = Hist2D(100, -3, 3, 100, 0, 6)
a.fill_array(np.random.multivariate_normal(
    mean=(0, 3),
    cov=np.arange(4).reshape(2, 2),
    size=(1E6,)))
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
#im = plt.contour(a, axes=ax1)
ax1.contour(a)
ax2.imshow(a)
plt.show()
```
