```python
from __future__ import division
def run(Nl,accl):
    '''
    + Caculates relative error in Acceptance given the number of Reconstructed Events(NR) and the Acceptance.
    + Relative error=f(NR,Acceptance)
    Input parameters:
    + Nl=List of NR
    + accl=List of Acceptances
    '''
    accRelErrl=[]
    errNl=[math.sqrt(x) for x in Nl]
    for acc in accl:
        Dl=[x/acc for x in Nl]
        errDl=[math.sqrt(x) for x in Dl]
        
        relErrNl=[errN/N for errN,N in zip(errNl,Nl)]
        relErrDl=[errD/D for errD,D in zip(errDl,Dl)]
        
        #Ql=[N/D for N,D in zip(Nl,Dl)]
        relErrQl=[math.sqrt(math.pow(relErrN,2)+math.pow(relErrD,2)) for relErrN,relErrD in zip(relErrNl,relErrDl)]
        accRelErrl.append(relErrQl)
    return accRelErrl    
```

```python
accl=[0.1,0.01,0.001]
Nl=[5,50,500,5000]
accRelErrl=run(Nl,accl)
#print accErrl

ax=subplot(111)
ax.set_title('Relative error in Acceptance vs. N(=Reconstructed events)')
cll=['r','b','g','k']
x=range(len(Nl))
xlabels=Nl
for i,accRelErr in enumerate(accRelErrl):
    ax.set_xticks(x)
    ax.set_xticklabels(xlabels)
    ax.scatter(x,accRelErr,c=cll[i],label='A=%s'%accl[i],s=50)
    ax.set_ylabel(r'$\frac{\delta a}{a}$',fontsize=30)
    ax.set_xlabel('N',fontsize=20)
ax.legend()
```
