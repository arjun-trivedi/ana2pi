```python
import rootnotes
```

```python
FST=ROOT.TFile(os.path.join(os.environ['DELASTDIR_SIM'],'siml','delastT.root'))
FSR=ROOT.TFile(os.path.join(os.environ['DELASTDIR_SIM'],'siml','delastR.root'))
FER=ROOT.TFile(os.path.join(os.environ['DELASTDIR_EXP'],'delastR.root'))
```

```python
h2_ST=FST.Get("delast/yield")
h2_SR=FSR.Get("delast/yield")
#! Calculate SA
h2_SA=h2_SR.Clone()
h2_SA.Divide(h2_ST)
#! Calculate SC
h2_SC=h2_SR.Clone()
h2_SC.Divide(h2_SA)
#! Calculate EC
h2_ER=FER.Get("delast/yield")
h2_EC=h2_ER.Clone()
h2_EC.Divide(h2_SA)
```

```python
ROOT.gStyle.SetOptStat(0)
c=rootnotes.default_canvas()
c.Divide(2,2)
c.cd(1)
h2_ST.Projection(0,1).Draw("colz")
c.cd(2)
h2_SR.Projection(0,1).Draw("colz")
c.cd(3)
h2_SA.Projection(0,1).Draw("colz")
c
```

```python
c1=rootnotes.default_canvas("c1")
c1.Divide(2,1)
c1.cd(1)
htheta_ST=h2_ST.Projection(0)
htheta_SC=h2_SC.Projection(0)
htheta_ST.SetLineColor(ROOT.gROOT.ProcessLine("kRed"))
htheta_SC.SetLineColor(ROOT.gROOT.ProcessLine("kBlue"))
htheta_ST.Draw()
htheta_SC.Draw("sames")
c1.cd(2)
htheta_SR=h2_SR.Projection(0)
htheta_ER=h2_ER.Projection(0)
htheta_SR.SetLineColor(ROOT.gROOT.ProcessLine("kRed"))
htheta_ER.SetLineColor(ROOT.gROOT.ProcessLine("kBlue"))
htheta_SR.Draw()
htheta_ER.Draw("sames")
c1
```

```python
c2=rootnotes.default_canvas("c2")
# htheta_SC=already obtained before
htheta_EC=h2_EC.Projection(0)
htheta_SC.SetLineColor(ROOT.gROOT.ProcessLine("kRed"))
htheta_EC.SetLineColor(ROOT.gROOT.ProcessLine("kBlue"))
htheta_SC.Draw()
htheta_EC.Draw("sames")
c2
```
