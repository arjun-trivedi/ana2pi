```python
import elaslib
import rootnotes
from rootpy.plotting import Hist, HistStack, Legend, Canvas
import rootpy.plotting.root2matplotlib as rplt
#nbins,xmin,xmax=46,14,60
def plot(be):
    #nbins,xmin,xmax=120,30,150
    nbins,xmin,xmax=46,14,60
    hThrtcl={}
    cold={'norad':'kRed','wrad':'kBlue'}
    for radtyp in ['norad','wrad']:
        hThrtcl[radtyp]=Hist(nbins,xmin,xmax,title='%s'%radtyp)
        #hThrtcl[radtyp].linecolor='green'
        #hThrtcl[radtyp].markercolor='green'
        hThrtcl[radtyp].SetLineColor(ROOT.gROOT.ProcessLine(cold[radtyp]))
        hThrtcl[radtyp].SetMarkerColor(ROOT.gROOT.ProcessLine(cold[radtyp]))
    for ibin in range(nbins):
        theta=hThrtcl['norad'].GetBinLowEdge(ibin+1)
        if theta==0: continue
        binc_norad=elaslib.elas(be,theta)
        binc_wrad=elaslib.elasrad(be,theta,2.5*0.00577,1.1)
        #print "theta,xsec=",theta,binc
        hThrtcl['norad'].SetBinContent(ibin+1,binc_norad)
        hThrtcl['norad'].SetBinError(ibin+1,0)
        hThrtcl['wrad'].SetBinContent(ibin+1,binc_wrad)
        hThrtcl['wrad'].SetBinError(ibin+1,0)
        #hThrtcl.GetXaxis().SetRangeUser(12,50)
    #c=rootnotes.default_canvas('c%s'%be,size=(500,500))
    #hThrtcl['norad'].Draw()
    #hThrtcl['wrad'].Draw("sames")
    hrto=hThrtcl['wrad'].Clone()
    hrto.SetTitle('#frac{wrad}{norad}')
    hrto.Divide(hThrtcl['norad'])
    fig=plt.figure(figsize=(7, 5), dpi=100)
    ax=plt.subplot(211)
    stack = HistStack()
    stack.Add(hThrtcl['norad'])
    stack.Add(hThrtcl['wrad'])
    rplt.errorbar(stack, xerr=False, emptybins=False, axes=ax)
    #rplt.errorbar(hThrtcl['wrad'],xerr=False, emptybins=False, axes=ax)
    ax.set_title("ep Cross section")
    ax.set_ylabel(r"$\frac{d\sigma}{d\Omega}[\mu b]$",fontsize=20)
    ax.set_yscale('log')
    ymin=minimum(elaslib.elas(be,100),elaslib.elasrad(be,100,2.5*0.00577,1.1))
    ax.set_ylim(ymin)
    plt.legend()
    ax=plt.subplot(212)
    rplt.errorbar(hrto, xerr=False, emptybins=False, axes=ax)
    ax.set_ylim(hrto.GetMinimum(),hrto.GetMaximum())#(0.80,0.82)
    #return hThrtcl['norad'],hThrtcl['wrad'],hrto
```

```python
plot(0.188)
```
