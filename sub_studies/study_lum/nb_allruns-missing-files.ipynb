```python
import glob#,subprocess
from subprocess import Popen, PIPE, STDOUT
from collections import OrderedDict
h10lstl=glob.glob("/home/trivedia/ongoing/e1f_allruns_h10lst/*.lst")
print "Number of Runs=",len(h10lstl)
runl=[]
nfl=[]
runs_missingd=OrderedDict()
for k in ['a00','a01','a02','a03','a04','a05','a06','a07','a08','a09',
          'a10','a11','a12','a13','a14','a15','a16','a17','a18','a19',
          'a20','a21','a22','a23','a24','a25','a26','a27','a28','a29',
          'a30','a31','a32','a33','a34','a35','a36','a37','a38','a39',
          'a41','a41','a42','a43','a44','a45','a46','a47','a48','a49']:
    runs_missingd[k]=[]
#{'a00':[],'a01':[],'a02':[],'a03':[],'a04':[],'a05':[]}
for h10lst in h10lstl:
    runnum=int(h10lst.split("/")[5].split(".")[0])
    #print runnum
    runl.append(runnum)
    f=open(h10lst)
    nfiles=len(f.readlines())
    nfl.append(nfiles)
    
#     f=open(h10lst) #reopen after readlines()
#     if not "a00" in f.readlines()[0]:
#         runs_missing_a00l.append(runnum)
        
    f=open(h10lst) #reopen after readlines()
    foundd=OrderedDict()
    for k in ['a00','a01','a02','a03','a04','a05','a06','a07','a08','a09',
              'a10','a11','a12','a13','a14','a15','a16','a17','a18','a19',
              'a20','a21','a22','a23','a24','a25','a26','a27','a28','a29',
              'a30','a31','a32','a33','a34','a35','a36','a37','a38','a39',
              'a41','a41','a42','a43','a44','a45','a46','a47','a48','a49']:
        foundd[k]=False
    #{'a00':False,'a01':False,'a02':False,'a03':False,'a04':False,'a05':False}
    for l in f.readlines():
        for k in foundd:
            if k in l: 
                foundd[k]=True
    for k in foundd:
        f_indx=int(k.split('a')[1])
        if foundd[k]==False and f_indx<nfiles:
            runs_missingd[k].append(runnum)
```
