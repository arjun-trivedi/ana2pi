```python
#! Read dlum.text and make displays
ff=open('dlum.txt')
runnuml=[]
norm_Ntrigger=[]
norm_Ngoode=[]
for l in ff.readlines():
    #print l
    dl=l.split(" ")
    runnum=int(dl[0].strip())
    q=float(dl[1].strip())
    Ntrigger=int(dl[2].strip())
    Ngoode=int(dl[3].strip())
    #print runnum,q,Ntrigger,Ngoode
    runnuml.append(runnum)
    if q==0.0:
        norm_Ntrigger.append(0)
        norm_Ngoode.append(0)
    else:
        norm_Ntrigger.append(Ntrigger/q)
        norm_Ngoode.append(Ngoode/q)
fig,axs=plt.subplots(figsize=(10,5),nrows=1,ncols=2)
axs[0].scatter(runnuml,norm_Ntrigger)
axs[1].scatter(runnuml,norm_Ngoode)
```
