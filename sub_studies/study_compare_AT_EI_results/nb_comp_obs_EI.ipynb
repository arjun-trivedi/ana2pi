```python
from __future__ import division
import rootnotes

#General global vars
SYST_ERR_EI=0.14 #14% taken from his analysis note

Q2L=[["2.00-2.40_1.400-2.125","cs20_24"],["2.40-3.00_1.400-2.125","cs24_30"],["3.00-3.50_1.400-2.125","cs30_35"],
     ["3.50-4.20_1.400-2.125","cs35_42"],["4.20-5.00_1.400-2.125","cs42_50"]]

#! get data to plot xsec vs Q2
#! R1 W=1.500-1.525
sgW1at,sgW1ei=[],[] #! 5th wbin
erW1at,erW1ei=[],[]
R1WBIN=5
#! R2 W=1.725-1.750 #! 14th wbin
sgW2at,sgW2ei=[],[]
erW2at,erW2ei=[],[]
R2WBIN=14

#! Store Q2 information: center,bin-width
#! + Calculated from Q2L
Q2C=[2.20,2.70,3.25,3.85,4.60]
Q2W=[0.20,0.30,0.25,0.35,0.40]

OUTDIR="/tmp/itg"
if not os.path.exists(OUTDIR):
	os.makedirs(OUTDIR)
    
for iq2,q2 in enumerate(Q2L):
    q2at=q2[0]
    q2ei=q2[1]
    #! get AT
    at_dir='%s/thesis_obs_norm_ST_shape_080317'%os.environ['OBSDIR_E162']
    #print at_dir
    if iq2+1<=2:
        fat=ROOT.TFile("%s/lowQ2/Obs_itg/Obs_itg.root"%at_dir,"READ")
    else:
        fat=ROOT.TFile("%s/highQ2/Obs_itg/Obs_itg.root"%at_dir,"READ")
    hat=fat.Get('%s/hW_obs_EF_THETA'%q2at)
    hat_err=fat.Get('%s/hW_err_EF_THETA'%q2at)
    #! Add syst-err to stat-err
    nbins=hat.GetNbinsX()
    for ibin in range(nbins):
        stat_err=hat.GetBinError(ibin+1)
        syst_err=hat_err.GetBinContent(ibin+1)
        #print stat_err, syst_err
        #! calculate total error and use that in hat
        tot_err=math.sqrt(stat_err**2+syst_err**2)
        hat.SetBinError(ibin+1,tot_err)
    #! get sgW1/W2 information
    sgW1at.append(hat.GetBinContent(R1WBIN))
    sgW2at.append(hat.GetBinContent(R2WBIN))
    erW1at.append(hat.GetBinError(R1WBIN))
    erW2at.append(hat.GetBinError(R2WBIN))
    
    #! get EI
    ei_dir='%s/EI_results'%os.environ['OBSDIR_E162']

    #! df version
    df=pd.read_csv('%s/%s.txt'%(ei_dir,q2ei),sep='\s{1,}',header=None,engine='python',thousands=',')
    #! Get number of rows (corresponds to number of wbins)
    nrows=len(df.index)
    #print df.head()
    #! create hei from hat.Clone and Reset
    hei=hat.Clone("hei")
    hei.Reset()
    hei.SetMarkerColor(1)
    hei.SetLineColor(1)
    #hei.SetMarkerStyle(2)
    for ibin in range(nrows):
        wbin,val,stat_err=df.loc[ibin,0],df.loc[ibin,1],df.loc[ibin,2]
        #print ibin+1,wbin,val,stat_err
        #! Calculate total error
        tot_err=math.sqrt(stat_err**2+(SYST_ERR_EI*val)**2)
        #! Now fill histogram
        hei.SetBinContent(ibin+1,val)
        hei.SetBinError(ibin+1,tot_err)
        #! get sgW1/W2 information
        if ibin+1==R1WBIN:  
            sgW1ei.append(val)
            erW1ei.append(tot_err)
        if ibin+1==R2WBIN:  
            sgW2ei.append(val)
            erW2ei.append(tot_err)

    #! find tuning
    #! +ymax
    if   q2at=="2.40-3.00_1.400-2.125":ymax=7.5
    elif q2at=="3.00-3.50_1.400-2.125":ymax=5.5
    if q2at=="2.40-3.00_1.400-2.125" or q2at=="3.00-3.50_1.400-2.125":
        hat.SetMaximum(ymax)

    cr=ROOT.TCanvas()
    hat.Draw()
    hei.Draw("same")
    #hei.Draw()
    #hat.Draw("same")
    cr.SaveAs("%s/c%s.png"%(OUTDIR,q2at))
    cr.SaveAs("%s/c%s.pdf"%(OUTDIR,q2at))
```

```python
from collections import OrderedDict
#! Note pad_map extended to include explicit variable name (as per EI's syntax)
pad_map=[(1,1,"M1","mpippr"),      (2,3,'M2',"mpimpr"),      (3,2,'M2','mpippim'),
	     (4,1,"THETA","theta_pim"),(5,3,'THETA','theta_pip'),(6,2,'THETA','theta_pr'),
		 (7,1,"ALPHA",'psi_pim'),  (8,3,'ALPHA','psi_pip'),    (9,2,'ALPHA','psi_pr')]
#! Set Q2WBIN in which comaparison is to be made
#! + AT and EI have different naming convention
Q2WBIN_DIR_AT="2.40-3.00_1.725-1.750"
Q2WBIN_DIR_EI="q2_24_30w_1725_1750"

#! Get AT data
fat=ROOT.TFile("%s/lowQ2/Obs_1D/Obs_1D.root"%at_dir,"READ")
hat=OrderedDict()
hat_err=OrderedDict()
for item in pad_map:
    pad,vst,var=item[0],item[1],item[2]
    #print pad,vst,var
    hat[vst,var]=fat.Get("%s/h1_EF_%d_%s"%(Q2WBIN_DIR_AT,vst,var))
    hat_err[vst,var]=fat.Get("%s/herr_h1_EF_%d_%s"%(Q2WBIN_DIR_AT,vst,var))
#! Add syst-err to stat-err
for k in hat:
    nbins=hat[k].GetNbinsX()
    for ibin in range(nbins):
        stat_err=hat[k].GetBinError(ibin+1)
        syst_err=hat_err[k].GetBinContent(ibin+1)
        #print stat_err, syst_err
        #! calculate total error and use that in hat
        tot_err=math.sqrt(stat_err**2+syst_err**2)
        hat[k].SetBinError(ibin+1,tot_err)
    
#! Get EI data
hei=OrderedDict()
for item in pad_map:
    pad,vst,var,varname=item[0],item[1],item[2],item[3]
    df=pd.read_csv('%s/%s/xsec1d_%s.dat'%(ei_dir,Q2WBIN_DIR_EI,varname),sep='\s{1,}',header=None,engine='python',thousands=',')
    #print df.head()
    hei[vst,var]=hat[vst,var].Clone()
    hei[vst,var].Reset()
    hei[vst,var].SetMarkerColor(1)
    hei[vst,var].SetLineColor(1)
    nrows=len(df.index)
    for ibin in range(nrows):
        binv,val,stat_err=df.loc[ibin,0],df.loc[ibin,1],df.loc[ibin,2]
        #print ibin+1,binv,val,stat_err
        #! Calculate total error
        tot_err=math.sqrt(stat_err**2+(SYST_ERR_EI*val)**2)
        #! Now fill histogram
        hei[vst,var].SetBinContent(ibin+1,val)
        hei[vst,var].SetBinError(ibin+1,tot_err)
        
#! draw together
c=ROOT.TCanvas("c","c",1000,1000)
pad_t=ROOT.TPad("pad_l","Legend pad",0.25,0.935,0.75,1.00)
pad_p=ROOT.TPad("pad_p","Plots pad",0.01,0.97,0.99,0.01)
pad_p.Draw()
pad_t.Draw()
pad_t.cd()
pt=ROOT.TPaveText(.05,.1,.95,.8)
pt.AddText("Q2_W bin=%s"%Q2WBIN_DIR_AT)
pt.SetTextSize(0.40)
pt.Draw()
pad_p.Divide(3,3)
for item in pad_map:
    pad,vst,var=item[0],item[1],item[2]
    #print "pad,vst,var=",pad,vst,var
    gpad=pad_p.cd(pad)

    #! fine tuning aesthetics
    #! + ymax
    if   var=="M1":    ymax=32
    elif var=="M2":    ymax=25
    elif var=="THETA": ymax=5
    elif var=="ALPHA": ymax=1.5
    hat[vst,var].SetMaximum(ymax)
    hat[vst,var].Draw()
    hei[vst,var].Draw("same")

    #! Add TLegend if pad==1
    if pad==1:
        l=ROOT.TLegend(0.60,0.70,0.90,0.90)
        l.SetBorderSize(1)
        l.SetTextSize(0.04)
        l.AddEntry(hat[vst,var],"AT","p")
        l.AddEntry(hei[vst,var],"EI","p")
        l.Draw()

#! Save canvas
outdir=("%s/%s"%("/tmp","1D"))
if not os.path.exists(outdir):
	os.makedirs(outdir)
c.SaveAs("%s/c.png"%(outdir))
c.SaveAs("%s/c.pdf"%(outdir))
```
